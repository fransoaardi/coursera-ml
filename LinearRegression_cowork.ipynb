```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def pplot(x, y):
    plt.xlabel('Population of city .. ')
    plt.ylabel('Profit')
    plt.scatter(x, y, c='r', marker='x')


df = pd.read_csv('data/ex1data1.txt', header=None, usecols=[0,1])
x = df.values[:, 0:1]
y = df.values[:, 1:2]
pplot(x, y)
```

```python
def computeCost(x, y, theta):
    m = x.size
    # J = (1/m)*x.transpose().dot(x.dot(theta)-y)
    J = (1/m)*(x.dot(theta)-y).transpose().dot(x.dot(theta)-y)
    return J


nx = np.c_[np.ones(x.size), x]
computeCost(nx, y, np.zeros((2, 1)))
```

```python
def gradientDescent(x, y, theta, alpha, num_iter):
    m = x.size
    j_hist = np.zeros(num_iter)
    for i in range(num_iter):
        j_hist[i] = computeCost(x, y, theta)
        theta = theta - alpha*(1/m)*x.transpose().dot(x.dot(theta)-y)
    return j_hist, theta
    
    
num_iter = 2000
j_hist, theta = gradientDescent(nx, y, np.zeros((2, 1)), 0.01, num_iter)
plt.plot(range(num_iter), j_hist)
print(theta)
```
